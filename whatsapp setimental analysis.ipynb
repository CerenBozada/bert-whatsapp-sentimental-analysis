```python
from transformers import AutoModelForSequenceClassification,AutoTokenizer,pipeline
model = AutoModelForSequenceClassification.from_pretrained(
    "savasy/bert-base-turkish-sentiment-cased")
tokenizer=AutoTokenizer.from_pretrained(
    "savasy/bert-base-turkish-sentiment-cased")
sentiment_analysis=pipeline(
    "sentiment-analysis",tokenizer=tokenizer,model=model)
```

```python
def sentiment_score(chat):
    analysis=sentiment_analysis(chat)[0]
    if analysis["label"] == "negative":
        return 0 - analysis["score"]
    return analysis["score"]
```

```python
with open("whatsapp.txt", encoding = 'utf-8') as f:
    lines = f.read().split("\n")[2:]
    lines=["".join(line.split("-")[1:]) for line in lines]
   
    lines=["".join(line.split(":")[1:]) for line in lines]
    lines=[line for line in lines if len(line) >3]
    lines=[line for line in lines if not "<Medya dahil edilmedi>" in line]

print(lines)  
```

```python
sentiments =[sentiment_score(line) for line in lines]
print (sentiments)
```

```python
import pandas as pd  
df = pd.DataFrame({"chat":lines, "sentiment":sentiments}) 
```

```python
df[df["sentiment"] > 0.9].sort_values(by=["sentiment"],ascending=False).head(50)
```
